# KDE Analysis of Asymmetry Plots

Using run (101) for the small counter test. 

```python
from MuonDataFrame import *
file1 = "processed_data/events_data_frame_101.h5"
## with bricks
mdfo = MuonDataFrame(file1, isNew=False, d1="last") #Muon Data Frame Object 
mdf = mdfo.events_df
```

```python
def getPD(term,df):
    df[term].plot(kind='kde')
    plt.title(term)
    plt.xlim([-0.5, 0.5])
    plt.show()
```

```python
def getAllPDs(df):
    plt.figure(1)
    
    plt.subplot(221)
    df['asymL1'].plot(kind='kde')
    plt.title('asymL1', fontsize=10)
    plt.xlim([-0.5, 0.5])
    
    plt.subplot(222)
    df['asymL2'].plot(kind='kde')
    plt.xlim([-0.5, 0.5])
    plt.title('asymL2', fontsize=10)
    
    plt.subplot(223)
    df['asymL4'].plot.kde(bw_method=0.05)
    plt.title('asymL3', fontsize=10)
    plt.xlim([-0.5, 0.5])
    
    plt.subplot(224)
    df['asymL4'].plot.kde(bw_method=0.05)
    plt.title('asymL4', fontsize=10)
    plt.xlim([-0.5, 0.5])
    plt.tight_layout()
    plt.show()
```

## Looking at KDE using all data

```python
getAllPDs(mdf)
```

### Extracting Fit Info from KDE distribution

```python
import statsmodels.api as sm
X = mdf['asymL1'].values
kde = sm.nonparametric.KDEUnivariate(X)
kde.fit()
```
